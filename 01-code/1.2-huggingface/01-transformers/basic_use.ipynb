```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

model_inputs = tokenizer(sequence)
model_inputs
```

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.",
            "So have I!"]

model_inputs = tokenizer(sequences)
model_inputs
```

```python
# Will pad the sequence up to the maximum sequence length of the sequences
model_inputs1 = tokenizer(sequences, padding="longest")

# Will pad the sequence up to the model max length (512 for BERT or DistilBERT)
model_inputs2 = tokenizer(sequences, padding="max_length")

# Will pad the sequences up to the specified max length
model_inputs3 = tokenizer(sequences, padding="max_length", max_length=8)

print(model_inputs1)
print(model_inputs2)
print(model_inputs3)
```

```python
# Will truncate the sequences that are longer than the model max length
model_inputs4 = tokenizer(sequences, truncation=True)
model_inputs5 = tokenizer(sequences, truncation=True, max_length=8)
print(model_inputs4)
print(model_inputs5)
```

```python
# returns PyTorch tensors
model_inputs6 = tokenizer(sequences, padding=True, return_tensors="pt")

# Returns NumPy arrays
model_inputs7 = tokenizer(sequences, padding=True, return_tensors="np")

print(model_inputs6)
print(model_inputs7)
```

```python
# special tokens

sequence = "I've been waiting for a HuggingFace course my whole life."

# tokenizer whole process
model_inputs = tokenizer(sequence)
print(model_inputs["input_ids"])

# tokenize segment the input sequence
tokens = tokenizer.tokenize(sequence)

# then convert_tokens_to_ids to encode the tokens
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
# why are they different ?
```

```python
print(tokenizer.decode(model_inputs["input_ids"])) # after tokenize
print(tokenizer.decode(ids)) # initial input
```
